```python
import pandas as pd 
import numpy as np 
import datetime as dt
import plotly.express as px
import seaborn as sns
sns.set(rc={'figure.figsize':(8,5)})
```

```python
df = pd.read_excel(r"C:\Users\Admin\Desktop\SCRAPY\amazon\Hamoye\Consolidated.xlsx")
```

```python
df.info()
```

```python
df.shape
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
df['Year'] = df['Date'].apply(lambda x:x.year)
```

```python
df['Year'] = df['Year'].fillna(0)
```

```python
# converting the year column to int
df['Year'] = df['Year'].astype(int)
```

```python
# selecting years that falls between 2009 and 2019
df = df[(df['Year'] >= 2009) & (df['Year'] <= 2019)]
```

```python
df['Year'].max()
```

```python
df['Year'].min()
```

```python
df.dtypes
```

```python
# extracting the ratings 
df['Ratings'] = df['Ratings'].str.split(" ", n = 1, expand = True)
```

```python
df['Ratings'].value_counts().head()
```

```python
# checking for missing values
df.isnull().sum()
```

```python
# filling missing values in ratings and reviews with zeros
df['Ratings'] = df['Ratings'].fillna(0)
df['Reviews_no'] = df['Reviews_no'].fillna(0)
```

```python
x = pd.DataFrame(df.loc[df['Year'] == 2009].groupby(['Year', 'Sub Category'])['Reviews_no'].sum().reset_index())
fig = sns.barplot( y ='Sub Category', x ='Reviews_no', data=x).set_title("2009 BOOKS BY RATING")
fig.figure.savefig('2009.png')
```
